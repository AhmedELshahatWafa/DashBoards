```python
import dash
import dash_core_components as dcc
import dash_html_components as html
import plotly.express as px
import dash_bootstrap_components as dbc
import pandas as pd

from dash_bootstrap_templates import load_figure_template
from dash.dependencies import Input,Output
```

```python
# This loads the "cyborg" themed figure template from dash-bootstrap-templates library,
# adds it to plotly.io and makes it the default figure template.

load_figure_template("cyborg")

app = dash.Dash(__name__, external_stylesheets=[dbc.themes.CYBORG])

df_data = pd.read_csv("../data/Video_Games_Sales_as_at_22_Dec_2016.csv")
# games sales for each region
fig1 = px.area(df_data.groupby("Year_of_Release").agg("sum"),labels={"Year_of_Release":"Year Of Release","value" : "Sales Value"}, y = ["NA_Sales","EU_Sales","JP_Sales","Other_Sales","Global_Sales"])
fig1.update_layout(
     margin = dict(l=20, r=20, t=30, b=50)
    ,width=860,height=335,
     title="Sales For Each Region",
    font=dict(
        family="Copperplate",
        size=14,
        color="LightGreen"
    )
)

year = 2015
df_data_filtered = df_data[df_data["Year_of_Release"] == year]
df_grouped = df_data_filtered.groupby("Platform").agg("sum")
df_grouped = df_grouped.sort_values("Global_Sales")
fig2=px.bar(df_grouped, labels={ "Global_Sales":"Global Sales", "Platform" : "Platform"}, x = "Global_Sales")
fig2.update_layout(
     margin=dict(l=20, r=20, t=30, b=20)
    ,width=600,height=620,
    title="Platforms Sales Over Years",
    font=dict(
        family="Copperplate",
        size=14,
        color="LightGreen"
    ))



df_genre = df_data.groupby("Genre").agg("sum")
fig3 = px.line(df_genre.sort_values("Global_Sales"), y="Global_Sales",markers="O", labels={"Global_Sales":"Global Sales"})
fig3.update_layout(
    margin=dict(l=20, r=140, t=30, b=10)
    ,width=860,height=335, 
    title="Heighest Sales Genre",
    font=dict(
        family="Copperplate",
        size=14,
        color="LightGreen"
    ))

df_data["mean_critic_score"] = df_data["Critic_Score"] / df_data["Critic_Count"]
df_genre_1 = df_data.groupby("Genre").agg("mean")
fig4= px.line(df_genre_1.sort_values("mean_critic_score"), y="mean_critic_score",markers="O")
fig4.update_layout(
    margin=dict(l=20, r=20, t=20, b=20)
    ,width=1500,height=500
)
print("x")
```
